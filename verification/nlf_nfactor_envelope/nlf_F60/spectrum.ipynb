```python
alphas = np.load('alphas.npy')
modes = np.load('modes.npy')
u = np.load('u.npy')
uy = np.load('uy.npy')
y = np.load('y.npy')
# want to plot mode nearest to alpha_r = 500 and alpha_i = 20 in a different color
# find the index of the mode closest to this value
alpha_target = 0.3 + 0 * 1.0j

distances = np.abs(alphas - alpha_target)
index = np.argmin(distances)
print(f"index = {index}, alpha = {alphas[index]}")

ylim_val = 0.2
plt.figure(figsize=(6,3),dpi=200)
plt.plot(np.real(alphas),np.imag(alphas),'o',color=colors[0])
plt.plot(np.real(alphas[index]),np.imag(alphas[index]),'o',color=colors[6])
plt.xlabel(r'$\Re(\alpha)$')
plt.ylabel(r'$\Im(\alpha)$')
plt.xlim([0.0, 1])
plt.ylim([-ylim_val, ylim_val])
plt.title(r'$\alpha = $' + f'{alphas[index]:.4f}')
plt.tight_layout()

# want to plot the mode associated with the target alpha 

plt.figure(figsize=(6,3),dpi=200)
plt.plot(np.abs(modes[:,index]), y)
plt.xlabel(r"$|u'|$")
plt.ylabel(r'$\eta$')
```
